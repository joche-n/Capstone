```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
name="k"
submissionlist=[]
index=0
while name!="p":


    index+=1
    score=input("enter the score as in 0.*** or 'f' for faild submission")
    contributer=input("who submitte the file 'j' for jochen 'l' for lara")
    name=input("press p if ther ar no more inputs")
    submissionlist.append((index,score,contributer))
    print(f"last entry was ({index}, {score}, {contributer})")

print(submissionlist)
```

```python
df=pd.DataFrame(submissionlist)
df.to_csv("submissions.csv",index=False)
```
